```python
import random

# Define inputs and desired outputs for scabies classification
x1 = [0, 0, 1, 1, 1, 1, 1]
x2 = [0, 1, 0, 0, 1, 1, 1]
x3 = [0, 1, 0, 1, 0, 1, 1]
x4 = [0, 1, 1, 0, 1, 0, 1]
y = [0, 1, 1, 0, 1, 0, 1]


w1, w2, w3, w4 = random.uniform(-1, 1), random.uniform(-1, 1), random.uniform(-1, 1), random.uniform(-1, 1)
theta = random.uniform(-1, 1)


def dot_product(w1, w2, w3, w4, x1, x2, x3, x4):
    return w1 * x1 + w2 * x2 + w3 * x3 + w4 * x4

def perceptron_output(w1, w2, w3, w4, x1, x2, x3, x4, theta):
    return 1 if dot_product(w1, w2, w3, w4, x1, x2, x3, x4) >= theta else 0

# Define a function to calculate accuracy
def calculate_accuracy(w1, w2, w3, w4, theta):
    correct = 0
    for i in range(len(x1)):
        predicted = perceptron_output(w1, w2, w3, w4, x1[i], x2[i], x3[i], x4[i], theta)
        if predicted == y[i]:
            correct += 1
    return correct / len(x1)

# Training parameters
learning_rate = 0.1
epochs = 1000

# Training the perceptron
for epoch in range(epochs):
    for i in range(len(x1)):
        predicted = perceptron_output(w1, w2, w3, w4, x1[i], x2[i], x3[i], x4[i], theta)
        if predicted != y[i]:
            if predicted == 0:
                w1 += learning_rate * x1[i]
                w2 += learning_rate * x2[i]
                w3 += learning_rate * x3[i]
                w4 += learning_rate * x4[i]
                theta -= learning_rate
            else:
                w1 -= learning_rate * x1[i]
                w2 -= learning_rate * x2[i]
                w3 -= learning_rate * x3[i]
                w4 -= learning_rate * x4[i]
                theta += learning_rate


    accuracy = calculate_accuracy(w1, w2, w3, w4, theta)

    y
    print(f"Epoch {epoch}:")
    print(f"Weights: w1={w1}, w2={w2}, w3={w3}, w4={w4}")
    print(f"Bias (): {theta}")
    print(f"Accuracy: {accuracy * 100}%")
    print('.')
    print('..')

    # Check for convergence (accuracy = 100%)
    if accuracy == 1.0:
        print("Training converged. 100% accuracy achieved.")
        break
```

```python
def standardize_with_maximum_norm(arr):
    if not arr:
        return arr
    max_norm = max(abs(x) for x in arr)

    # Standardize the array by dividing each element by the Maximum norm
    standardized_arr = [x / max_norm for x in arr]

    return standardized_arr
```

```python
def min_max_standardization(arr, new_min=0, new_max=1):
    if not arr:
        return arr
    min_val = min(arr)
    max_val = max(arr)

    # Standardize the array using Min-Max standardization formula
    standardized_arr = [(x - min_val) / (max_val - min_val) * (new_max - new_min) + new_min for x in arr]

    return standardized_arr
```

```python
import numpy as np

def z_score_standardization(arr):

    if not arr:
        return arr
    std_dev = np.std(arr)
    mean = np.mean(arr)

    # Standardize the array using Z-score standardization formula
    standardized_arr = [(x - mean) / std_dev for x in arr]

    return standardized_arr
```

```python
def standardize_inputs(input_data):
    mean = sum(input_data) / len(input_data)
    std_dev = (sum([(x - mean) ** 2 for x in input_data]) / len(input_data)) ** 0.5
    standardized_data = [(x - mean) / std_dev for x in input_data]
    return standardized_data
```

```python
def calculate_accuracy(predictions, labels):
    if len(predictions) != len(labels):
        raise ValueError("Predictions and labels must have the same length")

    correct_count = 0
    for i in range(len(predictions)):
        if predictions[i] == labels[i]:
            correct_count += 1

    return correct_count / len(predictions)

```

```python
def calculate_loss(predictions, labels):
    if len(predictions) != len(labels):
        raise ValueError("Predictions and labels must have the same length")

    loss = 0
    for i in range(len(predictions)):
        loss += (labels[i] - predictions[i]) ** 2

    return loss / (2 * len(predictions))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def train_perceptron(inputs, target_or, initial_weights):
    def dot_product(a, b):
        return np.dot(a, b)

    def activation_function(x):
        return 1 if x >= 0.5 else 0

    def calculate_accuracy(predictions, targets):
        return np.sum(predictions == targets) / len(targets)

    def squared_error_loss(y_true, y_pred):
        return np.mean((np.array(y_true) - np.array(y_pred))**2)

    weights = initial_weights.copy()
    epoch = 0

    accuracy_values = []
    loss_values = []

    while True:
        predictions = np.zeros(len(inputs))
        weight_updated = False

        for i in range(len(inputs)):
            net_input = dot_product(weights, inputs[i])
            prediction = activation_function(net_input)
            predictions[i] = prediction

            if prediction != target_or[i]:
                for j in range(len(weights)):
                    weights[j] += (target_or[i] - prediction) * inputs[i][j]
                weight_updated = True

        accuracy = calculate_accuracy(predictions, target_or)
        loss = squared_error_loss(target_or, predictions)

        accuracy_values.append(accuracy)
        loss_values.append(loss)

        print(f"Epoch {epoch + 1}:")
        print(f"Points: {inputs}")
        print(f"Desired Output: {target_or}")
        print(f"Calculated Output: {predictions}")
        print(f"Weights: {weights}")
        print(f"Updated: {weight_updated}")
        print(f"Accuracy: {accuracy:.2f}")
        print(f"Squared Error Loss: {loss:.4f}")

        if accuracy == 1.0:
            break

        epoch += 1

    # Plot confusion matrix
    cm = confusion_matrix(target_or, predictions)
    print("Confusion Matrix:")
    print(cm)


    # Plot loss vs. epoch graph
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, epoch + 2), loss_values, marker='o', linestyle='-', color='r')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss vs. Epoch')
    plt.grid(True)
    plt.show()
```

```python
# Define the training data
inputs = x_train
target_or = y_train

# Initialize weights and bias
initial_weights = np.array([0.0, 0.0, 0.0])

# Train the Perceptron
train_perceptron(inputs, target_or, initial_weights)
```

iteration

```python
import numpy as np

# Define the actual output (d)
d = np.array([0.2, 0.9])

# Initialize the initial weights and bias
w0 = -2
b0 = -2

# Set the learning rate ()
nu = 1.0

# Number of iterations
num_iterations = 50

# Define the input x
x = x

# Define lists to store the results after each epoch
epochs = []
weights = []
biases = []
y_preds = []
losses = []

# Define the sigmoid function
def sigmoid(x, w, b):
    exp = w * x + b
    y = 1 / (1 + np.exp(-exp))
    return y

# Define the loss function
def loss_function(y_true, y_pred):
    loss = np.sum((0.5) * ((y_pred - y_true) ** 2))
    return loss

# Implement the delta rule
for epoch in range(num_iterations):
    # Calculate the predicted output (y_pred)
    y_pred = sigmoid(x, w0, b0)

    # Calculate the loss
    loss = loss_function(d, y_pred)

    # Calculate the gradient for weight (dw) and bias (db)
    dw = np.sum((y_pred - d) * y_pred * (1 - y_pred) * x)
    db = np.sum((y_pred - d) * y_pred * (1 - y_pred))

    # Update the weights and bias
    w1 = w0 - nu * dw
    b1 = b0 - nu * db

    # Store the results for this epoch
    epochs.append(epoch + 1)
    weights.append(w1)
    biases.append(b1)
    y_preds.append(y_pred)
    losses.append(loss)

    # Update the weights and bias for the next epoch
    w0 = w1
    b0 = b1

# Print the results after each epoch in the specified format
print("Epoch\t     Weight\t                Bias\t                Predicted Y\t       Loss")
for e, w, b, y, l in zip(epochs, weights, biases, y_preds, losses):
    print(f"{e}\t{w:}\t{b:}\t{y:}\t{l:}")

# Print the final weights and bias
print("Final Weight :", w0)
print("Final Bias :", b0)
```

```python
# Create the iteration-versus-loss plot
import matplotlib.pyplot as plt
plt.plot(epochs, losses, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Iteration versus Loss')
plt.grid(True)
plt.show()
```
